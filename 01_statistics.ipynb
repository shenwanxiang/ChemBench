```python
from glob import glob
import pandas as pd
import os
```

```python
csvs = glob('./data_and_index/**/*.csv.gz')

res = []
for i in csvs:
    d = os.path.dirname(i)
    b = os.path.basename(d)
    task_name = b
    s = glob(os.path.join(d, '*.pkl'))
    df = pd.read_csv(i, compression = 'gzip')
    n, t = df.shape
    res.append([task_name, i, s, len(s), n, t-1]) 
    
df = pd.DataFrame(res, columns = ['task_name', 'data_file', 'split_file', 'n_cross_split', 'n_samples', 'n_task'])
```

```python
metric_dict =   {'ESOL': 'random, RMSE, regression, 01',
                 'FreeSolv': 'random,RMSE, regression, 02',
                 'Lipop': 'random,RMSE, regression, 03',
                 'PDBbind-full': 'time,RMSE, regression, 04',
                 'PDBbind-core': 'time,RMSE, regression, 05',
                 'PDBbind-refined': 'time, RMSE, regression, 06',
                 'PCBA': 'random,PRC_AUC, classification, 07',
                 'MUV': 'random,PRC_AUC, classification, 08',
                 'HIV': 'scaffold,ROC_AUC, classification, 09',
                 'BACE': 'scaffold,ROC_AUC, classification, 10',
                 'BBBP': 'scaffold,ROC_AUC, classification, 11',
                 'Tox21': 'random,ROC_AUC, classification, 12',
                 'ToxCast': 'random,ROC_AUC, classification, 13',
                 'SIDER': 'random,ROC_AUC, classification, 14',
                 'ClinTox': 'random,ROC_AUC, classification, 15',
                 'ChEMBL': 'scaffold,ROC_AUC, classification, 16'}
```

```python
dfx = pd.Series(metric_dict).apply(lambda x:x.split(',')).apply(pd.Series)
```

```python
dfx.columns = ['split_method', 'task_metrics', 'task_type', 'task_id']
```

```python
dff = df.set_index('task_name').join(dfx)
```

```python
dff = dff.sort_values('task_id').reset_index()
```

```python
x = ['task_name', 'task_id', 'task_type',  'n_samples', 'n_task', 'split_method', 
     'n_cross_split','task_metrics',  'data_file', 'split_file',]
dff = dff[x]
dff = dff.set_index('task_id')
```

```python
dff.to_pickle('summary.pkl')
```
